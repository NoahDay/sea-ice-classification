```python
# Import packages
import xarray as xr
import numpy as np
import math
import matplotlib.pyplot as plt
import csv
```

```python
with open('/Users/noahday/GitHub/sea-ice-classification/data/Heil2006/Circles.csv', newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        data_list = [row for row in csv_reader]

Sep_Nov = np.array(data_list, dtype=np.float64)

with open('/Users/noahday/GitHub/sea-ice-classification/data/Heil2006/Stars.csv', newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        data_list = [row for row in csv_reader]

Dec_Feb = np.array(data_list, dtype=np.float64)

with open('/Users/noahday/GitHub/sea-ice-classification/data/Heil2006/Crosses.csv', newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        data_list = [row for row in csv_reader]

Mar_May = np.array(data_list, dtype=np.float64)

with open('/Users/noahday/GitHub/sea-ice-classification/data/Heil2006/Dots.csv', newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        data_list = [row for row in csv_reader]

Jun_Aug = np.array(data_list, dtype=np.float64)
```

```python
plt.plot(Sep_Nov[:,0], Sep_Nov[:,1], label='Spring')
plt.plot(Sep_Nov[:,0], Dec_Feb[:,1], label='Summer')
plt.plot(Sep_Nov[:,0], Mar_May[:,1], label='Autumn')
plt.plot(Sep_Nov[:,0], Jun_Aug[:,1], label='Winter')
plt.legend()
plt.ylabel('Horizontal speed [m/s]')
```

```python
# Create a list of data arrays for each season
data = [Dec_Feb[:, 1], Mar_May[:, 1], Jun_Aug[:, 1], Sep_Nov[:, 1]]

# Labels for each season
seasons = ['Summer', 'Autumn', 'Winter', 'Spring',]

# Create a boxplot
plt.boxplot(data)

# Set xticks and labels
plt.xticks(range(1, len(seasons) + 1), seasons)

# Add labels and legend
plt.xlabel('Seasons')
plt.ylabel('Horizontal speed [m/s]')
plt.legend()

# Show the plot
plt.show()
```
